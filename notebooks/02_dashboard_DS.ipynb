```python
import pandas as pd
from pandasql import sqldf

```

```python

# Basic import
df = pd.read_parquet('../data/raw/eventsBenin.parquet')

# Specific engine (requires pyarrow or fastparquet)
# df = pd.read_parquet('data/raw/eventsBenin.parquet', engine='pyarrow')
```

```python

# Fonction pour executer le code SQL
pysqldf = lambda q: sqldf(q, globals())

# SQL test Liste des mois disponibles
query = "SELECT DISTINCT(MonthYear) AS 'Mois' FROM df ORDER BY MonthYear;"

result = pysqldf(query)
```

```python
result.sample()
```

```python
cameo_event_code = pd.read_csv('../data/raw/CAMEO.eventcodes.txt', sep='\t', dtype=str)
cameo = cameo_event_code.set_index("CAMEOEVENTCODE")["EVENTDESCRIPTION"].to_dict()
df["EventName"] = df["EventRootCode"].map(cameo).fillna("Other")
```

```python
pysqldf("SELECT EventName, COUNT(*) AS 'Nombre devenements' FROM df GROUP BY EventName ORDER BY 2 DESC")
```
